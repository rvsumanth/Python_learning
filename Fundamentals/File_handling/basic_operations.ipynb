# Basic Operations in File Handling


```python
# Read Data
f = open('demo.txt', 'r')
content = f.read()
print(content)
f.close()
```

```python
# Write Operation
f = open('demo.txt', 'w')
# existed data in the file is overridden with below text
f.write('Data in the text is over rided')
f.close()
```

```python
# Data Appending Operation
f = open('demo.txt', 'a')
f.write('\nThis is appended data to the existing data')
f.close()
```

```python
# Counting Number of words in a file
f = open('demo.txt', 'r')
text = f.read()
words = text.split()
print(f'Number of words in the file: {len(words)}')
f.close()
```

```python
# Reading only certain amount of data
f = open('demo.txt', 'r')
content = f.read(10) # Reads upto 10 indexes
print(content)
f.close()
```

```python
# Read line by line
f = open('demo.txt', 'r')
line1 = f.readline()
line2 = f.readline()
print(line1)
print(line2)
f.close()
```

```python
f = open('demo.txt', 'r')
line = f.readlines()
print(line)
f.close()
```

```python
# Using With statement
with open('demo.txt','r') as f:
    content = f.read()
    print(content)

# Using with statement we dont need to close file explicitly it closes automatically
```

```python
# Copy content from one file to another file
with open ('demo.txt', 'r') as f1, open ('demo2.txt', 'w') as f2:
    f2.write(f1.read()) 
    
```

```python
# Display only lines containing a keyword
keyword = 'data'
with open('demo.txt', 'r+') as f:
    for line in f:
        if keyword in line:
            print(line.strip())
```
